```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the directory containing the images
image_dir = '/content/data/000/'

# Get a list of all files in the directory
image_files = [os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.endswith('.png')]

# Display the first few images
for i, image_path in enumerate(image_files[:5]):  # Displaying only the first 5 images as an example
    img = mpimg.imread(image_path)
    plt.imshow(img)
    plt.title(os.path.basename(image_path))
    plt.axis('off')  # Hide axes
    plt.show()
```

```python
import os

hand_dir = "/content/data/"
hand_files = []

for folder_path, subfolders, filenames in os.walk(hand_dir):
    for filename in filenames:
        if filename.lower().endswith(".png"):
            hand_files.append(os.path.join(folder_path, filename))
```

```python
print(hand_files[:10])
```

```python
typed_dir='/content/test/resume'
typed_files=[]

for folder_path, subfolders, filenames in os.walk(typed_dir):
    for filename in filenames:
        if filename.lower().endswith(".tif"):
            typed_files.append(os.path.join(folder_path, filename))
```

```python
print(typed_files[:10])
```

label 0 for handwritten and 1 for typed

```python
handwritten_labels = [0] * len(hand_files)
typed_labels = [1] * len(typed_files)
```

```python
import pandas as pd
all_files = hand_files + typed_files
all_labels = handwritten_labels + typed_labels
df = pd.DataFrame({
    'filepath': all_files,
    'label': all_labels
})

# Shuffle the df
df = df.sample(frac=1).reset_index(drop=True)

print(df.head())
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
```

```python
df['label'] = df['label'].astype(str)

# Split the DataFrame into training (80%) and validation (20%) sets
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['label'])

#pixelsizw of images
IMG_SIZE = (350,350)
BATCH_SIZE = 32 #sendin in small grps so tat less press on ram

# Create a data generator for the training set with data augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,  #THESE ARE DATA AUGMENTATION STEPS-->SOTHAT THE MODEL DOESNT MEMORIZE THE LAYOUT OF THE ORIG IMGS
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    fill_mode='nearest'
)

#we do not aug the val data since we want unbiased checking
val_datagen = ImageDataGenerator(rescale=1./255.)

# Create the actual generator objects that will flow data from the DataFrame
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col='filepath',
    y_col='label',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary' # We have two classes
)

validation_generator = val_datagen.flow_from_dataframe(
    dataframe=val_df,
    x_col='filepath',
    y_col='label',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False # No need to shuffle validation data
)

```

```python
#pretrained on imagenet
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(350, 350, 3))

# Freeze the conv base model so its weights are not updated during training
base_model.trainable = False

# our FC layer after the base model
x = base_model.output
x = GlobalAveragePooling2D()(x) # Reduces dimensions
x = Dense(128, activation='relu')(x) # A fully connected layer
x = Dropout(0.5)(x) # Dropout for regularization to prevent overfitting
predictions = Dense(1, activation='sigmoid')(x) # The final output layer with 1 perceptron
# This is the final model we will train
model = Model(inputs=base_model.input, outputs=predictions)

# Compile the model, specifying the optimizer, loss function, and metrics
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

print("Model built and compiled successfully. Starting training...")

# Train the model
EPOCHS = 15  #but 2 WILL SUFFICE---->achieved good accuracy and less loss in 2nd epoch only
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator
)

print("\nModel training complete!")
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt

# --- 1. Define the Prediction Function ---
def predict_image(img_path, model):
    """
    Loads an image, preprocesses it, and predicts its class.
    """
    # Load the image from the path, resizing it to the model's expected input size
    img = image.load_img(img_path, target_size=(350, 350))

    # Convert the image to a NumPy array
    img_array = image.img_to_array(img)

    # The model expects input of shape (batch_size, height, width, channels)
    img_batch = np.expand_dims(img_array, axis=0)

    # Rescale the image pixel values just like the training data
    img_preprocessed = img_batch / 255.

    # Make the prediction
    prediction = model.predict(img_preprocessed)

    # model.predict() returns a value between 0 and 1 because of the sigmoid activation
    # We can set a threshold of 0.5 to decide the class
    if prediction[0][0] < 0.5:
        label = 'Handwritten'
    else:
        label = 'Typed'

    return label, img


# --- 2. Test the function on a few images from your validation set ---
# Get a few sample filepaths from your validation DataFrame `val_df`
sample_images = val_df.sample(4).to_dict('records')

# Loop through the sample images and predict
plt.figure(figsize=(12, 8))
for i, record in enumerate(sample_images):
    # Get the filepath and the true label
    filepath = record['filepath']
    true_label = 'Handwritten' if record['label'] == '0' else 'Typed'

    # Make the prediction
    predicted_label, img = predict_image(filepath, model)

    # Display the image and the prediction
    plt.subplot(2, 2, i + 1)
    plt.imshow(img)
    plt.title(f"True: {true_label}\nPredicted: {predicted_label}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
